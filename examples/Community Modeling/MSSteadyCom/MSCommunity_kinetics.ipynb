### variable kinetic coefficients

```python
modelInfo_2 = ["CMM_iAH991V2_iML1515.kb",40576]
mediaInfo_2 = ["Btheta_Ecoli_minimal_media",40576]

org_model = kbase.get_from_ws(modelInfo_2[0],modelInfo_2[1])
media = kbase.get_from_ws(mediaInfo_2[0],mediaInfo_2[1])
abundances = {"Species1":0.5,"Species2":0.5}

# evaluate the influence of kinetic constraints over a range of coefficients
%run ../../../modelseedpy/community/mscommunity.py
plots = []
for kinetic_coeff in range(800,1400,50):
    banner = '\n{}\nKinetic coefficient: {}'.format('='*50, kinetic_coeff)
    print(banner)
    
    model = org_model.copy()
    cfba = MSCommunity(model, abundances=abundances)
    cfba.gapfill(media)
    cfba.constrain(kinetic_coeff = kinetic_coeff)
    cfba.run(media)
    cross_feeding, exMets = cfba.compute_interactions(threshold=.1, filename=f"KinCoeff_{kinetic_coeff}.svg")
    display(cross_feeding)
    display(exMets)
```

### variable abundances

```python
modelInfo_2 = ["CMM_iAH991V2_iML1515.kb",40576]
mediaInfo_2 = ["Btheta_Ecoli_minimal_media",40576]
model = kbase.get_from_ws(modelInfo_2[0],modelInfo_2[1])
media = kbase.get_from_ws(mediaInfo_2[0],mediaInfo_2[1])
kinetic_coeff = 2000

# evaluate the influence of kinetic constraints over a range of coefficients
plots = []
for abundance in range(0, 10):
    species_1 = abundance/10   # B. Theta
    species_2 = 1-species_1    # E. coli
    banner = '\n{}\nAbundance\nspecies 1:{}\tspecies 2:{}'.format('='*30, species_1, species_2)
    print(banner)
    model = kbase.get_from_ws(modelInfo_2[0],modelInfo_2[1])
    media = kbase.get_from_ws(mediaInfo_2[0],mediaInfo_2[1])
    cfba = MSCommunity(model)
    cfba.drain_fluxes(media)
    cfba.gapfill(media)

    cfba.constrain(media, kinetic_coeff = kinetic_coeff, abundances = {"1":species_1,"2":species_2})
    solution = cfba.run()
    cfba.compute_interactions(solution)
    graph, pos, labels = cfba.visualize()
    plots.append((graph, pos, labels, banner))
```

### variable abundances

```python
modelInfo = ["electro.community.edited.v2",93204]
mediaInfo = ["CO2_minimal",93204]
model = kbase.get_from_ws(modelInfo[0],modelInfo[1])
media = kbase.get_from_ws(mediaInfo[0],mediaInfo[1])
kinetic_coeff = 2000

# evaluate the influence of kinetic constraints over a range of coefficients
plots = []
for abundance in range(0, 10):
    species_1 = abundance/10      # Acetobacterium
    species_2 = (1-species_1)/2   # Sulfurospirillum
    species_3 = (1-species_1)/2   # Desulfovibrio
    banner = '\n{}\nAbundance\nspecies 1:{}\tspecies 2:{}\tspecies 3:{}'.format('='*50, species_1, species_2,species_3)
    print(banner)
    
    model = kbase.get_from_ws(modelInfo[0],modelInfo[1])
    media = kbase.get_from_ws(mediaInfo[0],mediaInfo[1])
    cfba = MSCommunity(model)
    cfba.drain_fluxes(media)
    cfba.gapfill(media)
    
    cfba.constrain(media, kinetic_coeff = kinetic_coeff, abundances = {"1":species_1,"2":species_2, '3':species_3})
    solution = cfba.run()
    cfba.compute_interactions(solution)
    graph, pos, labels = cfba.visualize()
    plots.append((graph, pos, labels, banner))
```
